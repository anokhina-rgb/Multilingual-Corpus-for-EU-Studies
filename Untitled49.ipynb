```python
import re
import torch
from googletrans import Translator
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# ✅ Initialize Google translator
translator = Translator()

# ✅ Load a Hugging Face model for post-editing
# (you can change to another model if you want)
model_name = "Helsinki-NLP/opus-mt-en-uk"
print("⏳ Loading post-editing model…")
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
print("✅ Model loaded and ready!")
```

```python
def clean_text_for_translation(text: str) -> str:
    # Normalize and clean unwanted characters
    text = text.replace('\r\n', '\n').replace('\r', '\n')
    text = re.sub(r'[^\x20-\x7E\u00A0-\uFFFF]+', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

def split_into_chunks(text: str, max_size: int = 4500):
    # Split text into chunks within API limits
    chunks = []
    while len(text) > max_size:
        split_pos = text.rfind('.', 0, max_size)
        if split_pos == -1:
            split_pos = max_size
        chunks.append(text[:split_pos])
        text = text[split_pos:]
    if text:
        chunks.append(text)
    return chunks

def post_edit(text: str) -> str:
    """Run Hugging Face model to improve translation."""
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True).to(device)
    with torch.no_grad():
        outputs = model.generate(**inputs, max_length=512)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
from google.colab import files
from pathlib import Path

# ✅ Upload your TXT file
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
print(f"📄 File uploaded: {file_name}")

# ✅ Read file
with open(file_name, "r", encoding="utf-8", errors="ignore") as f:
    raw_text = f.read()

# ✅ Clean text
cleaned = clean_text_for_translation(raw_text)
chunks = split_into_chunks(cleaned)

# ✅ Translate and post-edit
translated_all = ""
for idx, ch in enumerate(chunks, start=1):
    print(f"🔄 Translating chunk {idx}/{len(chunks)}...")
    try:
        t = translator.translate(ch, dest='uk').text
    except Exception as e:
        print("⚠️ Google Translate error, using original text:", e)
        t = ch
    # Post-edit
    pe = post_edit(t)
    translated_all += pe + "\n"

# ✅ Save result
out_name = Path(file_name).stem + "_UA.txt"
with open(out_name, "w", encoding="utf-8") as f:
    f.write(translated_all.strip())

print("✅ Translation + post-editing done!")
files.download(out_name)
```
